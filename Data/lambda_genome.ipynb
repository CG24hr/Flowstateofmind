```python
def readgenome_fasta(filename) :

    """ read fasta file and turn them to string or dna sequence """

    genome = ''
    with open(filename, 'r') as f :
        for line in f :
            if line[0] in ['A', 'T', 'C', 'G'] : 
                genome += line.rstrip()
    return genome
genome = readgenome_fasta('lambda_virus.fa')
```

```python
print('lambda virus genome', genome)
print('number of bases :', len(genome))
```

```python
def naive(pattern, genome) : 

    """this function is naive exact matching algorithm 
       searching for a specific pattern within a larger genome sequence
    """ 

    positions = []
    for i in range(len(genome) - len(pattern) + 1) : 
        match = True
        for j in range(len(pattern)) : 
            if genome[i+j] != pattern[j] :
                match = False
                break
        if match :
            positions.append(i)
    return positions 
first_matches = naive('AGGT', genome)
second_matches = naive('ACCT', genome)

print("this first specific genome (AGGT) pattern found at {} positions : {}".format(len(first_matches), first_matches))
print("this second specific genome (ACCT) pattern found at {} positions : {}".format(len(second_matches), second_matches))
```

```python
matches = naive('TTAA', genome)
print("this specific genome (TTAA) pattern found at {} positions : {}".format(len(matches), matches))
```
